```python
import turtle as t

t.shape("turtle")
t.pensize(3)

#삼각형
t.color("red")
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)

#사각형
t.color("green")
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)

#오각형
t.color("blue")
t.forward(100)
t.left(72)
t.forward(100)
t.left(72)
t.forward(100)
t.left(72)
t.forward(100)
t.left(72)
t.forward(100)
t.left(72)
t.forward(100)
t.left(72)

# 왼쪽이동
t.penup()
t.goto(-100, 0)
t.pendown()

# 원
t.color("orange")
t.circle(100)

#별
t.color("brown")
t.forward(200)
t.left(144)
t.forward(200)
t.left(144)
t.forward(200)
t.left(144)
t.forward(200)
t.left(144)
t.forward(200)
t.left(144)



t.exitonclick()
```

```python
import turtle as t

t.shape("turtle")
t.pensize(3)

#삼각형
t.color("red")
for i in range(3):
    t.forward(100)
    t.left(120)

#사각형
t.color("green")
for i in range(4):
    t.forward(100)
    t.left(90)

#오각형
t.color("blue")
for i in range(5):
    t.forward(100)
    t.left(72)

# 왼쪽이동
t.penup()
t.goto(-100, 0)
t.pendown()

# 원
t.color("orange")
t.circle(100)

#별
t.color("brown")
for i in range(5):
    t.forward(200)
    t.left(144)



t.exitonclick()
```

```python
import turtle as t

t.shape("turtle")
t.pensize(1)
t.speed(10)

n = 50
t.bgcolor('black')
t.color('green')

for i in range(n):
    t.circle(80)
    t.left(360/n)
    
    

t.exitonclick()
```

```python
import turtle as t

t.shape("turtle")
t.pensize(1)
t.speed(0)

angle = 89

t.bgcolor('black')
t.color('yellow')

for i in range(10):
    t.fd(i)
    t.left(angle)
    
    

t.exitonclick()
```

```python
import turtle as t

t.shape("turtle")
t.pensize(1)
t.speed(0)
t.bgcolor('black')

for i in range(300):
    if i%3==0:
        t.color("red")
    elif i%3==1:
        t.color("blue")
    else:
        t.color("yellow")
        
    t.fd(i*2)
    t.left(119)
    
    

t.exitonclick()
```

```python
import turtle as t
import math as m

t.shape("turtle")
t.pensize(1)
t.speed(0)
t.bgcolor('black')
t.begin_fill()
t.color('red')

for i in range(100):
    h= m.pi*i/50
    i = 160*m.sin(h)**3
    j = 130*m.cos(h) - 50*m.cos(2*h) - 20*m.cos(3*h) - 10*m.cos(4*h)
    t.goto(i, j) 
t.end_fill()

t.exitonclick()
```
